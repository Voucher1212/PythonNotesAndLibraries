## distplot

The distplot shows the distribution of a univariate set of observations.
distplot is a versatile distribution plot in Seaborn that combines a histogram and a kernel density estimate (KDE) plot.
It allows you to visualize the distribution of a univariate (single variable) dataset. 
It displays a histogram with bars representing the frequency or count of observations and overlays a KDE curve to provide a smooth estimate of the underlying distribution. 

However, note that distplot has been deprecated since Seaborn version 0.11.0, and it is recommended to use displot or other specific plot functions instead.

```python
sns.distplot(tips['total_bill'])
```

```python
sns.histplot(tips['total_bill'])
```

## jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

### jointplot:

jointplot creates a scatter plot with two variables, along with marginal distribution plots on the axes. 
It allows you to visualize the relationship between two numeric variables and their individual distributions simultaneously. 
The central plot displays the scatter plot, while the top and right axes show the histograms or KDE plots of each variable separately. 
This plot is particularly useful for exploring bivariate relationships and identifying correlations between variables.

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

# pairplot:

* Pairplot is used to visualize pairwise relationships between multiple variables in a dataset. 
* It creates a grid of scatter plots or other types of plots to show the relationship between each pair of variables. 
* The diagonal of the plot grid typically displays histograms or KDE plots for each variable individually. 
* This plot is useful for gaining insights into the relationships and patterns among multiple variables in a dataset, especially in exploratory data analysis.

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sns.pairplot(data = tips)
```

```python
sns.pairplot(data = tips, hue ='sex',palette='rainbow')
```

```python
sns.pairplot(data = tips, hue ='day', palette = 'rainbow')
```

## rugplot

Rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:

* Rugplot creates a rug plot, which is a one-dimensional display of data points along an axis. 
* It adds small vertical lines or ticks to indicate the presence of individual data points. 
* Rug plots are often combined with other plots, such as KDE plots or histograms, to show the distribution of data points more comprehensively. 
* They are useful for visualizing the density and location of data points along an axis.

#### The rugplot function in Seaborn is primarily designed to display individual data points along an axis for continuous variables. It provides a visual representation of the density and location of data points. 

```python
sns.rugplot(tips['total_bill'])
```

```python
tips.info()
```

```python
sns.rugplot(tips['tip'])
```

# kdeplot

* kdeplot creates a kernel density estimate (KDE) plot, which visualizes the smoothed probability density function of a univariate dataset. 
* It estimates the underlying distribution of a variable by placing a kernel (a smoothed curve) at each data point and summing them to form the density estimate. 
* This plot is useful for visualizing the shape, peaks, and valleys of the distribution. 
* Multiple kdeplot calls can be used to compare distributions across different groups or categories.

```python
sns.kdeplot(data=tips, x="total_bill")
```

```python
import matplotlib.pyplot as plt
sns.kdeplot(data=tips, x="total_bill", bw_method=0.2, fill=True, color="blue")
plt.title("KDE Plot - Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")
```

## Barplot

The most basic categorical plot which is the bar plot.
A bar plot is just a general plot that allows you to aggregate the categorical data based off of some function and by default that's the mean you can almost think of this as a visualization of a group by action.

The black vertical lines on the bars in a bar plot are called "error bars." Error bars are graphical representations of the variability or uncertainty associated with the measurements or estimates represented by the bars.

Error bars provide a visual indication of the potential range or margin of error around the central value represented by the bar. They are commonly used to display confidence intervals, standard errors, or standard deviations associated with the data.

In a bar plot, error bars can be included to show the variability or uncertainty in the data points being represented by each bar. The length of the error bars indicates the degree of uncertainty or variability.

```python
# barplot and countplot
#These very similar plots allow you to get aggregate data off a categorical feature in your data. 
# barplot is a general plot that allows you to aggregate the categorical data based off some function, 
# by default the mean
sns.barplot(x='sex',y='total_bill',data=tips,palette='rainbow')
```

```python
# And we can change what is known as the estimator object if you want to basically affect
# what the actual grouping or aggregation here is.
# adding another argument here estimator.

import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std,palette='rainbow')
```

### countplot

This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences. 
Which is why we only pass the x value:

```python
sns.countplot(data=tips, x="day", palette = 'rainbow')
plt.xlabel("Day of the Week")
plt.ylabel("Count")
plt.title("Count of Tips by Day")
plt.show()
```

And this will create a bar plots and by default what it's showing you here is the average or mean of the total bill.

```python
sns.barplot(data=tips, x="day", y="total_bill",palette='rainbow')
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
# Calculate the average total bill by day
average_total_bill = tips.groupby("day")["total_bill"].mean()

# Create a bar plot with average total bill on the y-axis
sns.barplot(x=average_total_bill.index, y=average_total_bill.values,palette='rainbow')
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill")
plt.title("Average Total Bill by Day")
```

## boxplot and violinplot

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x='day',y='total_bill', data=tips,palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue ='smoker',palette='rainbow')
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

## Violin Plot

Now the violin plot unlike the box plot allows us to actually plot all the components that correspond

to actual data points and it's essentially showing the kernel density estimation of the underlying distribution

```python
sns.violinplot(x='day',y='total_bill', data=tips,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, hue = 'sex' , split ='true',palette='rainbow')
```

## stripplot and swarmplot
The stripplot will draw a scatterplot where one variable is categorical. 
A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. 
This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='rainbow')
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue ='sex')
```

```python
### Combining Categorical Plots
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

# Matrix Plots

Matrix plots in Python are used to visualize the relationships between multiple variables in a tabular form. They allow us to examine the pairwise relationships between variables and identify patterns or correlations.

Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data (later in the machine learning section we will learn how to formally cluster data).

There are different types of matrix plots available in Python, and two commonly used ones are the correlation matrix plot and the heatmap.

Correlation Matrix Plot:
A correlation matrix plot displays the correlation coefficients between pairs of variables. It provides a visual representation of the strength and direction of the linear relationship between variables.
Heatmap:
A heatmap is a graphical representation of data in a tabular form, where the values are represented using colors. It is particularly useful for visualizing matrices or tables with numerical values.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights.head()
```

## Heatmap

In order for a heatmap to work properly, your data should already be in a matrix form, the sns.heatmap function basically just colors it in for you. For example:

```python
tips.head()
```

```python
# Matrix form for correlation data
tips.corr()

# The correlation coefficients range from -1 to 1,
# with values closer to 1 indicating a stronger positive correlation, 
# values closer to -1 indicating a stronger negative correlation, 
# and values close to 0 indicating a weak or no correlation.
```

```python
tips.corr(numeric_only=True)
```

```python
sns.heatmap(tips.corr(numeric_only=True))
```

```python
sns.heatmap(tips.corr(numeric_only=True),cmap='coolwarm',annot=True)
```

```python
# Load a dataset
data = sns.load_dataset("iris")

# Compute the correlation matrix
# The correlation matrix is a useful tool for 
# understanding the relationships between variables in a dataset,
# identifying patterns, and performing exploratory data analysis.

corr_matrix = data.corr(numeric_only= 'True')

# Create the correlation matrix plot
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix Plot")
plt.show()
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(pvflights)
```

### In Python, the "pivoting" or "pivoting table" operation refers to reshaping a DataFrame by converting values from one column into multiple columns, creating a tabular structure. 

### This operation is commonly used for summarizing and aggregating data.

```python
sns.heatmap(pvflights,cmap='magma',linecolor='white',linewidths=1)
```

## clustermap

The clustermap uses hierarchal clustering to produce a clustered version of the heatmap. For example:

```python
sns.clustermap(pvflights)
```

Notice now how the years and months are no longer in order, instead they are grouped by similarity in value (passenger count). That means we can begin to infer things from this plot, such as August and July being similar (makes sense, since they are both summer travel months)

```python
# More options to get the information a little clearer like normalization
sns.clustermap(pvflights,cmap='coolwarm',standard_scale=1)
```

# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

We will see Seaborn's grid capabilities to automate subplots from features in our data.

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
#We can map a specific type
pg = sns.PairGrid(iris)
pg.map(plt.scatter)
```

```python
#We can map a specific type
pg = sns.PairGrid(iris)
pg.map(sns.kdeplot)
```

```python
# Map to upper,lower, and diagonal
#gives a lot more control
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips = sns.load_dataset('tips')
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill')
tips.head()
```

```python
tips = sns.load_dataset('tips')
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill','tip')
```
