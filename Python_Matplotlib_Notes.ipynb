## Matplotlib's object oriented api methods

* We will instantiate figure objects and then call methods or attributes from that object.
* The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. 
* This approach is nicer when dealing with a canvas that has multiple plots on it. 
* To begin we create a figure instance. Then we can add axes to that figure:

```python
# Matplotlib's object oriented api methods

# creating a figure object, its like a blank canvas

fig = plt.figure()

# Adding set of axes to the figure object
 # pass a list to create axes
    # list takes 4 arguments ; left, bottom , width and height
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # pass a list to create axes

# plotting on those axes

axes.plot(x,y)

axes.set_xlabel('x axis') # Notice the use of set_ to begin methods
axes.set_ylabel('y axis')
axes.set_title('My Object Oriented Python Plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.4,0.4,0.5,0.5])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.5,0.5])

axes2 = fig.add_axes([0.2,0.3,0.5,0.5])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.set_title('Larger Plot')
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
# call fig,axes using plt.subplots()
fig, axes = plt.subplots()
# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# Then you can specify the number of rows and columns when creating the subplots() object:
# Empty canvas of 1 by 2 subplots
# tupple unpacking
fig, axes = plt.subplots(nrows=3, ncols=3)
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4)
plt.tight_layout() #spaces them out
```

```python
# Axes is an array of axes to plot on axes
```

```python
# subplots is a front handle on plt.figure . its like an axes manager
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)

axes[0].set_title('First Plot') 

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

## Figure size, aspect ratio and DPI
### Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 


```python
fig = plt.figure(figsize=(3,2),dpi= 100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
#using figsize

fig = plt.figure(figsize=(8,2),dpi= 100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
# using subplots
fig,axes  = plt.subplots(nrows= 2, ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig,axes  = plt.subplots(nrows= 1, ncols=2, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

## Setting colors, linewidths, linetypes

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake:

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color ='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color ='#FF8C00') # RGB Hex code 
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color ='#FF8C00', linewidth=4, alpha =0.25)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color ='purple', lw = 3, linestyle ='-.', alpha =0.5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color ='purple', lw = 3, marker ='o', alpha =0.5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color ='cyan', lw = 3, marker ='+', markersize = 40 ,alpha =0.5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, x+16, color="blue", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

## Control over axis appearance

In this section we will look at controlling axis sizing properties in a matplotlib figure.

## Plot range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, 
or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color ='green')
ax.set_xlim(0,2)
ax.set_ylim(0,2)
```
